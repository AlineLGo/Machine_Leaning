```python
import cv2 #OpenCV
import numpy as np # bibilioteca que trabalha com matriz
from matplotlib import pyplot as plt # trabalha com gr√°ficos
from google.colab.patches import cv2_imshow # mostrar as imagens utilizando OpenCV
```

```python
img = cv2.imread('/content/drive/MyDrive/Curso de Machine Learning/paisagem.jpg')
plt.imshow(img) # traz a imagem no formato BGR
```

```python
# Fazendo a convers√£o da imagem para RGB
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb) # inverte os valores para cada pixel
```

```python
# Transformando para tons de cinza
gray = cv2.cvtColor(rgb, cv2.COLOR_RGB2GRAY)
plt.imshow(gray, cmap='gray')
```

```python
limiar = 100 # 0 -255
```

```python
val, thresh = cv2.threshold(gray, limiar, 255, cv2.THRESH_BINARY)
```
