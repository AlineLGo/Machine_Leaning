```python
# fazendo a transformação do tipo do dado "portatil" da planilha
cleanup = {"Portatil": {"Smartphone": 1, "Tablet": 2}}
```

```python
data.replace(cleanup, inplace=True) # aqui está fazendo alteração dos valores do dataframe original
data
```

```python
# aqui acontece a separação das colunas com a coluna Portatil a qual se deseja fazer a predição
y= data.Portatil # coluna resposta
X= data.drop(columns =['Portatil']) # vai selecionar todas as informações das outras colunas, menos da coluna Portatil
```

# Aqui que começa o treinamento da maquina para diferenciar Tablet de Smartphone

```python
from sklearn.tree import DecisionTreeClassifier # importando a biblioteca scikitlearn, dentro dela importando um classificador de arvore de decisão
from sklearn.model_selection import train_test_split # atraves da biblioteca está importando um modelo de treino, separando uma porcentagem para treino e outra porcentagem para teste

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4)
# X_train é a parte do algoritmo que vai entregar para a maquina aprender
# X_test é outra parte do algoritmo para ser aplicado o teste depois do aprendizado
# y_train é o gabarito dos dados de treino (X_train) para saber o quanto a maquina acertou
# y_test é o gabarito dos dados de teste (X_test)
#test_size significa que vai tirar um tamanho dos dados para depois serem aplicados nos testes, nesse caso 0.4 ou 40% dos dados vai ser usados para o teste e 60% vai ser usado para o treinamento
clf = DecisionTreeClassifier() # aqui recebe o objeto da arvore de decisão
clf.fit(X_train, y_train) # aqui vai receber a função de aprendizado (os dados de treino) sendo aqui que a maquina faz o aprendizado.
```

```python
y_pred = clf.predict(X_test) # aqui é a variavel resposta do algoritmo, passando o teste
y_true = y_test # recebendo o gabarito dos testes

print("Resultados obtidos: ")
print(y_pred)
print("\n Resultados esperados: ")
print(y_true.values)
```

```python
from sklearn import metrics # importando as metricas para verificar a precisão do modelo
print("Precision" + str(metrics.precision_score(y_true, y_pred)))
# no resultado abaixo, mostra que teve uma taxa de precisão de 100%
```
